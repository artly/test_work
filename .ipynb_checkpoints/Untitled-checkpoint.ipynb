```python

# вспомогательная функция для парсинга тегов fb2
# во второй версии Питона так лучше не делать, но это 3я
def parse_p(node):
    words = []
    if list(node):

        # Лучше это писать через try ... except, но я боюсь, что try .. except TypeError
        # может замаскировать непредвиденные ошибки
        if node.text:
            words += [[t, 0] for t in re.split('\s+', node.text)]

        for sub in list(node):

            if sub.tag == 'a':
                pass
            elif sub.tag == 'emphasis':
                # выделенный текст
                words += [[t, 1] for t in re.split('\s+', sub.text)]

            else:
                print(sub.tag)
                print(sub.text)

            # то же самое, что выше: читается лучше с try, но с точки зрения fail fast лучше так
            if sub.tail:
                words += [[t, 0] for t in re.split('\s+', sub.tail)]

    else:
        words += [[t, 0] for t in re.split('\s+', node.text)]
    return words

# рекурсивно обходим вложенные разделы в fb2-файле, достаём названия глав
def word_table(node, title):
    #print('node = {}'.format(node.tag))
    word_list = []
    table_list = []
    subtitle_text = None
    words = pd.DataFrame(data={'word': [],
                               'chapter': [],
                               'is_emphasized': []})
    for el in list(node):
        if el.tag == 'section':
            subtitle = el.find('title/p')

            try:
                subtitle_text = subtitle.text
            except AttributeError:
                pass
            if title:
                subtitle_text = title + '. ' + subtitle_text
            else:
                subtitle_text = subtitle_text

            #print(subtitle_text)

            table_list += word_table(el, subtitle_text)
        elif el.tag == 'p':
            word_list += parse_p(el)


        elif el.tag == 'poem':
            for v_tag in el.find('./stanza/v'):
                word_list += [[w, 0] for w in re.split('\s+', v_tag.text)]
        elif el.tag == 'cite':
            if el.find('./p'):
                for p_tag in el.find('./p'):
                    word_list += parse_p(p_tag)
            else:
                try:
                    for subtitle_tag in el.find('./subtitle'):
                        word_list += [[w, 0] for w in re.split('\s+', subtitle_tag.text)]
                except TypeError:
                    print('В теге cite что-то не так')
                    pass
    if word_list:
        res = pd.DataFrame(data=np.array(word_list), columns=['word', 'is_emphasized'])
        res['chapter'] = title
        return [res]
    else:
        return table_list
```

```python
# открываем файл, парсим xml
path = 'p+n.fb2'
tree = ET.parse(path)
body = [child for child in tree.findall('./') if 'body' in child.tag and not child.attrib][0]

# собираем и объединяем главы в единый df 
raw_table = pd.concat(word_table(body, None)).reset_index(drop=True)
```

```python
end_sentence = re.compile('[\.\!\?»…]+')
start_sentence = re.compile('«')
dialog_break = '–'

# удаляем переносы строк
raw_table.drop(raw_table[raw_table['word']==''].index, inplace=True)

# выделим и пронумеруем предложения
raw_table['sentence_n'] = (raw_table['word']
 .apply(lambda x: 1 if re.search(end_sentence, x) else 0)
 .shift(1)
 .fillna(0))
raw_table.loc[0,'sentence_n'] = 1
raw_table.reset_index(drop=True, inplace=True)
raw_table.loc[raw_table[(raw_table['word'] == dialog_break) &
                        (raw_table['sentence_n'] > 0)].index + 1, 'sentence_n'] += 1

#считаем предложения из диалогов
raw_table['dialog'] = 0
raw_table.loc[raw_table[(raw_table['word'] == dialog_break) & 
                        (raw_table['sentence_n'] > 0)].index + 1, 'dialog'] = 1

# подчищаем теперь уже лишние тире в начале предложений
raw_table.drop(raw_table[raw_table['word'] == dialog_break].index, inplace=True)
raw_table.reset_index(drop=True, inplace=True)

raw_table['sentence'] = raw_table['sentence_n'].cumsum()

# посчитаем запятые и точки с запятыми
raw_table['comma'] = raw_table['word'].str.contains(',').astype(int)
raw_table['semicolon'] = raw_table['word'].str.contains(';').astype(int)

# посчитаем знаки экспрессии
raw_table['exclamation'] = raw_table['word'].str.contains('!').astype(int)
raw_table['q_mark'] = raw_table['word'].str.contains('\?').astype(int)
raw_table['ddd'] = raw_table['word'].str.contains('…|\?\.\.|!\.\.').astype(int)

# добавим номера глав
raw_table['chapter_n'] = ((raw_table['chapter'].shift(1) != raw_table['chapter'])
                  .astype(int)
                  .cumsum())
# двойные знаки пунктуации разделялись на 2 строчки. Когда мы посчитали все знаки, можно удалить такие строки из 1го знака пунктуации
raw_table = (raw_table.drop(raw_table[raw_table['word'].str.match(r'\W$')].index)
             .reset_index(drop=True))

#
# пунктуацию посчитали, можно убирать 
raw_table['word'] = raw_table['word'].str.replace(r'(^\W+)|(\W+$)', '')

# находим имена собственные. 
# Re плох с заглавными буквами в юникоде, я проявляю смекалку, чтобы не писать сложных выражений 
raw_table['is_capital'] = (raw_table['word'].str.slice(stop=1) == 
                           raw_table['word'].str.slice(stop=1).str.upper())
raw_table['is_proper'] = (raw_table['is_capital'] & (raw_table['sentence_n'] == 0))

# теперь можно нормализовать регистр
raw_table['word'] = raw_table['word'].str.lower()

# выделим слова русского языка
raw_table['is_russian'] = raw_table['word'].str.contains(r'[А-Яа-яЁё]')

# некоторые нерусские слова - цифры
raw_table['is_digit'] = raw_table['word'].str.isdigit()

# длина слов и количество гласных
raw_table['length'] = raw_table['word'].apply(len)
raw_table['vowels'] = raw_table['word'].apply(lambda x: len(re.findall(r'[аоиеяюыэё]', x)))

# насладимся результатом
words_clean = raw_table[['word', 'sentence', 'chapter_n', 'length', 'is_digit', 'is_russian', 
                 'is_proper', 'is_emphasized', 'vowels']]

# отдельно статистика по предложениям
sentences_clean = (raw_table
                         .groupby(['sentence'])['comma', 'semicolon', 'exclamation', 'q_mark', 'ddd']
                         .apply(np.sum)
                         .merge(raw_table
                         .groupby(['sentence']).agg({'word': 'count', 'length': 'mean'}), 
                                left_index=True, right_index=True)
                      )
```

```python
words_total = words_clean.shape[0]
words_rus = words_clean[(~(words_clean['is_digit']) & words_clean['is_russian'])].shape[0] / words_total
words_in_sentence = sentences_clean['word'].mean()
sentences_total = sentences_clean.shape[0]
sentences_per_chapter = words_clean[['sentence', 'chapter_n']].drop_duplicates().groupby('chapter_n').agg('count')['sentence'].mean()
words_per_chapter = words_clean.groupby('chapter_n').count()['word'].mean()
chapters_total = words_clean['chapter_n'].drop_duplicates().shape[0]
punctuation_per_sentence = (sentences_clean['comma'] + sentences_clean['semicolon']).sum() / sentences_total
sentiment_per_sentence = (sentences_clean['exclamation'] + sentences_clean['q_mark'] 
                          + sentences_clean['ddd']).sum() / sentences_total
exclamations = (sentences_clean['exclamation']).sum() / sentences_total
q_marks = (sentences_clean['q_mark']).sum() / sentences_total
ddd = (sentences_clean['ddd']).sum() / sentences_total
```
